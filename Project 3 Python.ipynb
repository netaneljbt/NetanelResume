```python
#rename columnes and save the change in the dataset : 
cof.rename(columns={'transaction_id':'OrderId'},inplace=True)
cof.rename(columns={'transaction_date':'OrderDate'},inplace=True)
cof.rename(columns={'transaction_time':'OrderTime'},inplace=True)
cof.rename(columns={'transaction_qty':'OrderQty'},inplace=True)
cof.rename(columns={'unit_price':'unit_price($)'},inplace=True)
## dateset update check:
```

```python
#change check:
cof.head()
```

```python
# Reviewing column 'product_detail' , info is not relevant in my opinion and will be drop:
cof['product_detail'].unique()
```

```python
#drop column 'product_detail' from dataset:
cof.drop(columns='product_detail',inplace=True)
```

```python
# Drop check:
cof.info()
```

```python
#Creating additional dates columns (week,month,quarter,year),make sure OrderDate type = datetime64:
```

```python
#creat Week column :
cof['Week']=cof['OrderDate'].dt.isocalendar().week
```

```python
#creat Month column :
cof['Month']=cof['OrderDate']
```

```python
# fromat to month only:
cof['Month']=cof['Month'].dt.month
```

```python
#creat Quarter column :
cof['Quarter']=cof['OrderDate']
```

```python
cof['Quarter']=cof['Quarter'].dt.quarter
```

```python
#creat year column:
cof['Year']=cof['OrderDate'].dt.year
```

```python
#new columns info check:
cof.info()
```

```python
# dateset update check:
cof.head()
```

```python
#change 'OrderTime' type to datetime in order to creat an hour column:

cof['OrderTime']=pd.to_datetime(cof['OrderTime'],format='%H:%M:%S')
```

```python
#creat Hour column:
cof['Hour']=cof['OrderTime']
```

```python
cof['Hour']=cof['Hour'].dt.hour
```

```python
# new columns check:
cof.info()
```

```python
# dateset update check:
cof.head()
```

```python
#creat new column 'Sales' :
cof['Sales']= cof['unit_price($)']*cof['OrderQty']
```

```python
# dateset update check:
cof.head()
```

```python
# gruping to new dataset the 'product_category' and 'store_location' by sales:
pro_store_sales=cof.groupby(['product_category','store_location'])['Sales'].sum().sort_values(ascending=False).reset_index(name='Sales')
```

```python
#crat new column 'part_of_day' Dividing the day hours into 3 parts: morning, noon and evning (I used Copilot guidance for this section):
def cat_time(hour):
    if hour <12:
        return 'Morning'
    elif hour <18:
        return 'Noon'
    else:
        return 'Evning'

cof['part_of_day']=cof['Hour'].apply(cat_time)
```

```python
# countplot to visualize total orderqty by 'part of day':
plt.figure(figsize=(12,4))
sns.countplot(data=cof,x='store_location',hue='part_of_day')
plt.title('total orderqty by part of day')
plt.xlabel('part of day')
plt.ylabel('total orderqty')
plt.show()
```
